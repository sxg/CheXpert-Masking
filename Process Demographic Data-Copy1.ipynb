```python
import pandas as pd
import os
import re
import numpy as np
```

```python
df = pd.read_excel("~/Documents/CheXpert-v1.0-small/CHEXPERT DEMO.xlsx")
```

```python
df.shape
```

```python
train_df = pd.read_csv("~/Documents/CheXpert-v1.0-small/train.csv")
valid_df = pd.read_csv("~/Documents/CheXpert-v1.0-small/valid.csv")
```

```python
train_df.shape
```

```python
valid_df.shape
```

```python
df.head()
```

```python
train_df.head()
```

```python
train_df["PATIENT"] = train_df["Path"].apply(lambda x: re.search(r"patient\d+", x).group())
valid_df["PATIENT"] = valid_df["Path"].apply(lambda x: re.search(r"patient\d+", x).group())
train_df.head()
```

```python
train_df = pd.merge(train_df, df, on="PATIENT", how="left")
valid_df = pd.merge(valid_df, df, on="PATIENT", how="left")
train_df.shape
```

```python
train_df.head()
```

```python
train_df["PRIMARY_RACE"].value_counts()
```

```python
mask = (train_df["PRIMARY_RACE"].str.contains("Black", na=False))
train_df.loc[mask, "race"] = "BLACK"

mask = (train_df["PRIMARY_RACE"].str.contains("White", na=False))
train_df.loc[mask, "race"] = "WHITE"

mask = (train_df["PRIMARY_RACE"].str.contains("Asian", na=False))
train_df.loc[mask, "race"] = "ASIAN"

mask = (valid_df["PRIMARY_RACE"].str.contains("Black", na=False))
valid_df.loc[mask, "race"] = "BLACK"

mask = (valid_df["PRIMARY_RACE"].str.contains("White", na=False))
valid_df.loc[mask, "race"] = "WHITE"

mask = (valid_df["PRIMARY_RACE"].str.contains("Asian", na=False))
valid_df.loc[mask, "race"] = "ASIAN"
```

```python
train_df.shape
```

```python
train_df.head()
```

```python
valid_df.shape
```

```python
valid_df.head()
```

```python
train_df["White"] = train_df["race"].apply(lambda x: 1 if x == "WHITE" else 0)
train_df["Black"] = train_df["race"].apply(lambda x: 1 if x == "BLACK" else 0)
train_df["Asian"] = train_df["race"].apply(lambda x: 1 if x == "ASIAN" else 0)

valid_df["White"] = valid_df["race"].apply(lambda x: 1 if x == "WHITE" else 0)
valid_df["Black"] = valid_df["race"].apply(lambda x: 1 if x == "BLACK" else 0)
valid_df["Asian"] = valid_df["race"].apply(lambda x: 1 if x == "ASIAN" else 0)
```

```python
train_df.head(15)
```

```python
races = ["WHITE", "BLACK", "ASIAN"]
train_df = train_df.query("race in @races")
valid_df = valid_df.query("race in @races")
train_df.head()
```

```python
valid_df.shape
```

```python
train_df = train_df.drop_duplicates(subset="PATIENT", keep="first")
valid_df = valid_df.drop_duplicates(subset="PATIENT", keep="first")
```

```python
print(train_df.shape)
```

```python
print(valid_df.shape)
```

```python
total_df = pd.concat([train_df, valid_df], axis=0)
total_df.head()
```

```python
total_df.shape
```

```python
from sklearn.model_selection import train_test_split

train_df, valid_df = train_test_split(total_df, train_size=0.8, random_state=48)
```
