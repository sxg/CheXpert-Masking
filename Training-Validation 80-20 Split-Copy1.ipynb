```python
import pandas as pd
import os
import re
import numpy as np
```

```python
demo_df = pd.read_excel("~/Documents/CheXpert-v1.0-small/CHEXPERT DEMO.xlsx")
```

```python
demo_df.shape
```

```python
train_df = pd.read_csv("~/Documents/CheXpert-v1.0-small/train.csv")
valid_df = pd.read_csv("~/Documents/CheXpert-v1.0-small/valid.csv")
```

```python
train_df.shape
```

```python
valid_df.shape
```

```python
df = pd.concat([train_df, valid_df], axis=0)
```

```python
df.shape
```

```python
df.head()
```

```python
df["PATIENT"] = df["Path"].apply(lambda x: re.search(r"patient\d+", x).group())
df.head()
```

```python
df = pd.merge(df, demo_df, on="PATIENT", how="left")
df.shape
```

```python
df.head()
```

```python
df["PRIMARY_RACE"].value_counts()
```

```python
mask = (df["PRIMARY_RACE"].str.contains("Black", na=False))
df.loc[mask, "race"] = "Black"

mask = (df["PRIMARY_RACE"].str.contains("White", na=False))
df.loc[mask, "race"] = "White"

mask = (df["PRIMARY_RACE"].str.contains("Asian", na=False))
df.loc[mask, "race"] = "Asian"
```

```python
df.shape
```

```python
df.head()
```

```python
df_one_hot = pd.get_dummies(df["race"], dtype=int, prefix='', prefix_sep='')
df = pd.concat([df, df_one_hot], axis=1)
df.shape
```

```python
df.head(25)
```

```python
filtered_df = df[(df[["White", "Black", "Asian"]].any(axis=1)) & (df["Frontal/Lateral"] == "Frontal")].drop_duplicates(subset="PATIENT", keep="first")
filtered_df.shape
```

```python
filtered_df["race"].value_counts()
```

```python
N = 3099  # 3099 unique black patients in the filtered dataset, get the same number of Asian and white patients
dfs = [group.head(N) for _, group in filtered_df.groupby("race")]
final_df = pd.concat(dfs, ignore_index=True)
```

```python
final_df["race"].value_counts()
```

```python
from sklearn.model_selection import train_test_split

train_df, valid_df = train_test_split(final_df, train_size=0.8, random_state=42)
```
